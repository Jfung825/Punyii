## âœ… Practice 1: Define a Class

Create a class `Book` with attributes `title`, `author`, and `pages`. Add a method `summary()` that prints: 
`<title> by <author>, <pages> pages`

```python
class NPC :
    def __init__(self,name,hp,mana):
        self.name = name 
        self.hp = hp
        self.mana = mana 
        self.status = "LIVED"
    
    def STAT(self) :

        if self.hp <= 0 :
            self.status = "DEAD"
            self.hp = 0

        print(f"\nName : {self.name} ({self.status})\nHP : {self.hp}\nMANA : {self.mana}")

    def DMG(self,damage):
        self.hp -= damage
        print(f"{self.name} -{damage} hp")

    def ATTACK(self,damage,npc) :
        npc.hp -= damage
        print(f"{self.name} attacked {npc.name} -{damage} hp")

p1 = NPC("Fluke",20,15)      
p2 = NPC("SPN HK",30,10)

p1.ATTACK(500,p2)
p2.STAT()
```
